# Quizaufgaben

1. Wie können wir die Heatmap jetzt so einstellen, dass wir nur "starke" Korrelationen sehen ("stark":= PCC>0.7)
2. Gibt es gefährliche Trends über der Zeit?

```python
import pandas
import numpy
import seaborn
import matplotlib.pyplot as plt

t = pandas.read_csv("Berlin_crimes.csv")
t2 = t.iloc[:, 4:]

M = t2.corr()

# فیلتر کردن فقط همبستگی‌های قوی (PCC > 0.7)
strong_corr = M[M.abs() > 0.7]

# نقشه حرارتی فقط برای همبستگی‌های قوی
plt.figure(figsize=(10, 8))
seaborn.heatmap(strong_corr, cmap="coolwarm", mask=numpy.triu(strong_corr))
plt.title("Heatmap of Strong Correlations (PCC > 0.7)")
plt.show()
```

```python
# بررسی اینکه آیا ستون 'Date' وجود دارد و به فرمت datetime تبدیل می‌شود
if 'Date' in t.columns:
    t['Date'] = pandas.to_datetime(t['Date'])
    
    # استخراج سال از تاریخ
    t['Year'] = t['Date'].dt.year
    
    # شمارش تعداد جرایم در هر سال
    yearly_trends = t.groupby('Year').size()
    
    # نمایش روند جرایم در طول زمان
    plt.figure(figsize=(10, 6))
    yearly_trends.plot(kind='line', marker='o')
    plt.title('Yearly Trend of Crimes')
    plt.xlabel('Year')
    plt.ylabel('Number of Crimes')
    plt.grid(True)
    plt.show()
else:
    print("\n'Error: The dataset does not contain a 'Date' column.")
```

```python
# 1. Datensatz aus dem WWW runterladen
import pandas
t = pandas.read_csv("http://www.juergenbrauer.org/kurse/daten/Berlin_crimes.csv")

# 2. Nur die Deliktanzahlen in Tabelle t2 "slicen"
t2 = t.iloc[:, 4:]

# 3. Alle paarweisen Pearson-Korrelationskoeffizienten berechnen
M = t2.corr()

# 4. Visualisierung der Koeffizienten durch Farben ("Heatmap")
import numpy
import seaborn
import matplotlib.pyplot as plt
plt.figure(figsize=(9,9))
seaborn.heatmap( M, cmap="coolwarm", mask=numpy.triu(M), annot=True, fmt=".1f" )
plt.title("Korrelationstabelle für Delikte in Berlin")
plt.show()
```
